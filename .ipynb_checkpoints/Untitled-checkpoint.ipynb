```python
import pandas as pd
import numpy as np

regions = [
    "Краснодарский край","Ставропольский край","Ростовская обл.","Волгоградская обл.",
    "Саратовская обл.","Воронежская обл.","Белгородская обл.","Курская обл.",
    "Орловская обл.","Тамбовская обл.","Липецкая обл.","Пензенская обл.",
    "Ульяновская обл.","Самарская обл.","Оренбургская обл.","Челябинская обл.",
    "Свердловская обл.","Тюменская обл.","Омская обл.","Новосибирская обл.",
    "Алтайский край","Красноярский край","Иркутская обл.","Бурятия",
    "Забайкальский край","Республика Татарстан","Башкортостан","Чувашская Респ.",
    "Марий Эл","Мордовия","Удмуртская Респ.","Пермский край",
    "Нижегородская обл.","Ивановская обл.","Ярославская обл.","Тверская обл.",
    "Смоленская обл.","Брянская обл.","Рязанская обл.","Тульская обл."
]

years = [2019, 2020, 2021, 2022, 2023]

np.random.seed(42)
rows = []
for i, region in enumerate(regions):
    base = i / len(regions)
    for yi, year in enumerate(years):
        trend = yi * 0.03
        kfh = round(2 + base * 8 + np.random.uniform(0, 0.5) + trend, 2)
        coops = int(5 + base * 60 + np.random.uniform(0, 5) + yi * 1.5)
        lpx = int(120 + base * 800 + np.random.uniform(0, 50) + yi * 15)
        income = int(12000 + base * 18000 + np.random.uniform(0, 1000) + yi * 500)
        unemp = round(4 + (1 - base) * 8 + np.random.uniform(0, 0.8) - trend * 0.5, 1)
        outflow = round(max(0, 15 - kfh * 0.4 - coops * 0.05 - (income / 10000) * 0.3 + unemp * 0.6 + np.random.uniform(-2, 2)), 2)
        rows.append([region, year, outflow, kfh, coops, lpx, income, unemp])

df = pd.DataFrame(rows, columns=[
    'region', 'year', 'outflow_per1000', 'kfh_per1000',
    'coops_count', 'lpx_mln_rub', 'avg_income_rub', 'unemployment_pct'
])

df.to_csv('dataset_agro_migration.csv', index=False, encoding='utf-8-sig')
print(f"Датасет создан: {len(df)} строк")
df.head()
```

```python
import statsmodels.api as sm

X = df[['kfh_per1000', 'coops_count', 'lpx_mln_rub', 'avg_income_rub', 'unemployment_pct']]
y = df['outflow_per1000']

X_const = sm.add_constant(X)
model = sm.OLS(y, X_const).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(df['coops_count'], df['outflow_per1000'], alpha=0.5, color='steelblue')
axes[0].set_xlabel('Количество кооперативов')
axes[0].set_ylabel('Отток населения (на 1000 чел.)')
axes[0].set_title('Кооперативы vs Отток')

axes[1].scatter(df['unemployment_pct'], df['outflow_per1000'], alpha=0.5, color='coral')
axes[1].set_xlabel('Безработица, %')
axes[1].set_ylabel('Отток населения (на 1000 чел.)')
axes[1].set_title('Безработица vs Отток')

plt.tight_layout()
plt.savefig('graphs.png', dpi=150)
plt.show()
```
